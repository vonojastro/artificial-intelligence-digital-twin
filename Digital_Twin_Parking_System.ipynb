**Phase 1 (Park In)**

```python
from datetime import datetime

def process_vehicle_transaction(vehicle_df, transaction, vehicle, parking):
    try:
        # Randomize vehicle table
        random_row = rand_vehicle('F')
        # select top most row that has A
        top_most_row = get_top_most_row_with_status_A(get_parking_data_as_dataframe())

        # Get current date and time
        current_datetime = datetime.now()
        date_in = current_datetime.strftime("%Y-%m-%d")
        time_in = current_datetime.strftime("%H:%M:%S")

        # Insert data into transaction collection
        inserted_id = transaction.insert_one({
            'Plate Number': random_row['Plate Number'].values[0],
            'Parking Location': top_most_row['Parking Location'],
            'Parking Area': int(top_most_row['Parking Area']),
            'Date in': date_in,
            'Time in': time_in,
            'Date out': '',
            'Time out': '',
        }).inserted_id

        if inserted_id:
            # Update 'Status' to 'P' in the 'vehicle' collection
            vehicle.update_one({'Plate Number': random_row['Plate Number'].values[0]}, {'$set': {'Status': 'P'}})
            # Update 'Status' to 'N' in the 'parking' collection
            parking.update_one({'Parking Location': top_most_row['Parking Location'], 'Parking Area': int(top_most_row['Parking Area'])}, {'$set': {'Status': 'N'}})

            print("Park In Successfully")
        else:
            print("Failed to process transaction.")
    except Exception as e:
        print("An error occurred:", str(e))
```

**Phase 2 (Park Out)**

```python
def update_transaction_and_clock_out(transaction, vehicle, parking):
    try:
        x = rand_vehicle('P')
        plate_number = x['Plate Number'].iloc[0]

        # Search transaction data
        query = {'Plate Number': plate_number, 'Date out': '', 'Time out': ''}
        matched_transactions = list(transaction.find(query))

        # Get current date and time
        current_datetime = datetime.now()
        date_out = current_datetime.strftime("%Y-%m-%d")
        time_out = current_datetime.strftime("%H:%M:%S")

        # Update 'Date out' and 'Time out' fields in the transaction collection
        if matched_transactions:
            update_query = {'$set': {'Date out': date_out, 'Time out': time_out}}
            transaction.update_many(query, update_query)

            # Get parking location and area from the matched transaction
            matched_transaction = matched_transactions[0]
            parking_location = matched_transaction['Parking Location']
            parking_area = matched_transaction['Parking Area']

            # Update 'Status' to 'F' in the vehicle collection
            vehicle.update_one({'Plate Number': plate_number}, {'$set': {'Status': 'F'}})
            # Update 'Status' to 'A' in the parking collection
            parking.update_one({'Parking Location': parking_location, 'Parking Area': parking_area}, {'$set': {'Status': 'A'}})
            print("Park Out Successfully")
        else:
            print("No matching transactions found.")
    except Exception as e:
        print("An error occurred:", str(e))
```

**Randomize Function**

```python
import random
number = [1,2]
def randomize_and_process(number, vehicle_df, transaction, vehicle, parking):
    try:
        # Randomly choose between 1 and 2
        choice = random.choice(number)
        if choice == 1:
            vehicleStatus = vehicle.count_documents({'Status': 'P'})
            vehicle_count = vehicle.count_documents({})
            if vehicleStatus == vehicle_count:
                clock_out = update_transaction_and_clock_out(transaction, vehicle, parking)
                return clock_out
            else:
            # Process vehicle transaction (Clock In)
                clock_in = process_vehicle_transaction(vehicle_df, transaction, vehicle, parking)
                return clock_in
        elif choice == 2:
            # Check if any parking has 'N' value in status
            parking_with_N = parking.count_documents({'Status': 'N'})
            if parking_with_N > 0:
                # Update transaction and clock out
                clock_out = update_transaction_and_clock_out(transaction, vehicle, parking)
                return clock_out
            else:

                clock_in = process_vehicle_transaction(vehicle_df, transaction, vehicle, parking)
                return clock_in
        else:
            print("Invalid choice.")
    except Exception as e:
        print("An error occurred:", str(e))
```
